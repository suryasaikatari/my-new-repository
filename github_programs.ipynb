```python
a = 10
b = 20
print(a + b)
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print(a + b)
```

```python
with open("add.txt", "w") as f:
    f.write("10 20")
with open("add.txt") as f:
    a, b = map(int, f.read().split())

print(a + b)
```

```python
with open("add_multi.txt", "w") as f:
    f.write("2 3\n4 5\n6 7")
with open("add_multi.txt") as f:
    for line in f:
        a, b = map(int, line.split())
        print(a + b)
```

program 4

```python
n = 5
fact = 1
for i in range(1, n + 1):
    fact *= i
print(fact)
```

```python
n = int(input("Enter number: "))
fact = 1
for i in range(1, n + 1):
    fact *= i
print(fact)
```

```python
with open("fact.txt", "w") as f:
    f.write("5")
with open("fact.txt") as f:
    n = int(f.read())
fact = 1
for i in range(1, n + 1):
    fact *= i
print(fact)
```
